## Data visualisation

### Outliers detection and treatment

```python
plt.figure(figsize=(16, 9))
sns.violinplot(x="loudness",inner='quartile', data=df)
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.boxplot(x="loudness",  data=df,color='g')
plt.show()
```

```python
#finding inter quartile range(iqr)
percentile25 = df['loudness'].quantile(0.25)
percentile75 = df['loudness'].quantile(0.75)
iqr = percentile75-percentile25
```

```python
#finding upper and lower limit
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
print(upper_limit)
print(lower_limit)
```

```python
#finding outlier
df[df['loudness'] > upper_limit]
df[df['loudness'] < lower_limit]
```

```python
#trimming( here only loudness is  taken in consideration)
df = df[(df['loudness'] <= upper_limit) & (df['loudness'] >= lower_limit)]
```

```python
#now again creating box plot to show that there is no outlier
plt.figure(figsize=(16, 9))
sns.boxplot(x="loudness",  data=df, color='r')
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.violinplot(x="loudness", inner='quartile', data=df, color='y')
plt.show()
```
